In the module Alexey works with a csv file and uses pandas to infer some of the schema, after that it was repartitioned and once repartitioned it was saved as a parquet file. I already started working with a parquet file (also from NYC taxi data). Repartition of a DF is possible no matter whether it came from a CSV, or parquet file.

```python
df = df.repartition(24)
df.write.mode("overwrite").parquet("fhvhv/2025/01/")
```

```python
#reading the new parquet files
df = spark.read.parquet("fhvhv/2025/01/")
df.show()
df.printSchema()


```
